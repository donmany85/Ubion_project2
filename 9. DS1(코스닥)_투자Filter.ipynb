# 1. 3년 연속 순이익이 증가한 기업

### 1-1 코스피

```python
# 데이터 로드
ni_kospi = pd.read_csv("./datasets/kospi_영업이익관련(종진형님).csv", encoding='cp949')
ni_kosdaq = pd.read_csv("./datasets/kosdaq_영업이익관련(종진형님).csv", encoding='cp949')
```

```python
# 코스피 컬럼명 변경
ni_kospi.columns = ['회사명', '거래소코드', '회계년도', '매출총이익(연결)', '당기순이익(연결)', '영업이익(연결)',
       '매출총이익(개별)', '당기순이익(개별)','영업이익(개별)']
```

```python
# 결측치 확인
ni_kospi.isna().sum()
```

```python
# 연결
ni_kospi_ifrs = ni_kospi[['매출총이익(연결)', '당기순이익(연결)', '영업이익(연결)']]
ni_kospi_ifrs.columns = [['매출총이익', '당기순이익', '영업이익']]
ni_kospi_ifrs
```

```python
# 개별
ni_kospi_gaap = ni_kospi[['매출총이익(개별)', '당기순이익(개별)','영업이익(개별)']]
ni_kospi_gaap.columns = [['매출총이익', '당기순이익', '영업이익']]
ni_kospi_gaap
```

```python
# 회사명 거래소코드 회계년도 따로 저장
ni_kospi_info = ni_kospi[['회사명', '거래소코드', '회계년도']]
```

```python
# 연결 결측치 개별로 채우는 코드
ni_kospi_combined = ni_kospi_ifrs.combine_first(ni_kospi_gaap)
ni_kospi_combined.isna().sum()
```

```python
kospi_revenue = pd.concat([ni_kospi_info, ni_kospi_combined], axis=1)
kospi_revenue.columns = ['회사명','거래소코드','회계년도','매출총이익','당기순이익','영업이익']
kospi_revenue
```

### 1-2 코스닥

```python
# 컬럼명 변경
ni_kosdaq.columns = ['회사명', '거래소코드', '회계년도', '매출총이익(연결)', '당기순이익(연결)', '영업이익(연결)',
       '매출총이익(개별)', '당기순이익(개별)','영업이익(개별)']
```

```python
ni_kosdaq.isna().sum()
```

```python
ni_kosdaq_ifrs = ni_kosdaq[['매출총이익(연결)', '당기순이익(연결)', '영업이익(연결)']]
ni_kosdaq_ifrs.columns = [['매출총이익', '당기순이익', '영업이익']]
ni_kosdaq_ifrs
```

```python
ni_kosdaq_gaap = ni_kosdaq[['매출총이익(개별)', '당기순이익(개별)','영업이익(개별)']]
ni_kosdaq_gaap.columns = [['매출총이익', '당기순이익', '영업이익']]
ni_kosdaq_gaap
```

```python
ni_kosdaq_info = ni_kosdaq[['회사명', '거래소코드', '회계년도']]
```

```python
ni_kosdaq_combined = ni_kosdaq_ifrs.combine_first(ni_kosdaq_gaap)
ni_kosdaq_combined
```

```python
kosdaq_revenue = pd.concat([ni_kosdaq_info, ni_kosdaq_combined], axis=1)
kosdaq_revenue.columns = ['회사명','거래소코드','회계년도','매출총이익','당기순이익','영업이익']
```

```python
revenue = pd.concat([kospi_revenue, kosdaq_revenue], axis=0)
```

```python
revenue.drop_duplicates(inplace=True)
```

```python
revenue.drop(labels='매출총이익', axis=1, inplace=True)
```

```python
# 결측치 제거
revenue.dropna(inplace=True)
revenue
```

```python
revenue['3년당기순이익흑자'] = np.where((revenue['당기순이익'] > 0) &
                            (revenue['당기순이익'].shift(1) > 0) &
                            (revenue['당기순이익'].shift(2) > 0), 1, 0)
```

```python
revenue['3년영업이익흑자'] = np.where((revenue['영업이익'] > 0) &
                            (revenue['영업이익'].shift(1) > 0) &
                            (revenue['영업이익'].shift(2) > 0), 1, 0)
```

```python
revenue[revenue['3년당기순이익흑자'] == 1]
```

```python
revenue.reset_index(drop=True, inplace=True)
```

```python
revenue
```

# 3. 시가총액 붙이기

```python
시가총액 = pd.read_csv('./datasets/kospi+kosdaq_시가총액.csv', index_col=0)
시가총액
```

```python
# 종목코드를 6자리로 통일. ex) 990 -> 000990
def padding(x):
    x = str(x)
    return x.rjust(6, "0")
```

```python
filtered['거래소코드'] = filtered['거래소코드'].map(padding)
```

```python
merge_df = pd.merge(filtered, 시가총액, how='left', on=['거래소코드', '회계년도'])
merge_df
```

```python
merge_df.drop(columns=['종가','상장주식수'], axis=1, inplace=True)
```

```python
merge_df
```

```python
merge_df.to_csv('./datasets/투자/KOSDAQ_DATA(투자).csv')
```
