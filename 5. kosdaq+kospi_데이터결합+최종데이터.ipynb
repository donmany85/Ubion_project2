```python
import pandas as pd
```

```python
kospi = pd.read_csv('./datasets/4.kospi_더미화완료.csv', index_col=0)
kosdaq = pd.read_csv('./datasets/4.kosdaq_더미화완료.csv', index_col=0)
```

```python
kospi
```

```python
kosdaq
```

# 데이터 결합

```python
df_concat = pd.concat([kospi,kosdaq], axis=0)
```

```python
df_concat.reset_index(drop=True,inplace=True)
```

```python
df_concat.info()
```

# 결합과정 중 중복데이터 제거 코드

```python
df_concat.drop_duplicates(subset=['거래소코드', '회계년도'], keep='first', inplace=True)
```

```python
df_concat.reset_index(drop=True,inplace=True)
```

```python
df_concat.to_csv('./datasets/KOSDAQ+KOSPI_DATA.csv')
kosdaq.to_csv('./datasets/KOSDAQ_DATA.csv')
```

# Kospi Data Spilt

```python
# 회계년도 기준 데이터 스플릿 진행
df_concat_1117 = df_concat[df_concat['회계년도'] <= '2017/12']
df_concat_1819 = df_concat[(df_concat['회계년도'] <= '2019/12') & (df_concat['회계년도'] >= '2018/12')]
df_concat_2021 = df_concat[(df_concat['회계년도'] <= '2021/12') & (df_concat['회계년도'] >= '2020/12')]
```

```python
df_concat_1117['label'].value_counts()
```

```python
df_concat_1819['label'].value_counts()
```

```python
# 인덱스 초기화
df_concat_1117.reset_index(drop=True, inplace=True)
df_concat_1819.reset_index(drop=True, inplace=True)
df_concat_2021.reset_index(drop=True, inplace=True)
```

```python
df_concat_1117['label'].value_counts()
```

```python
df_concat_1819['label'].value_counts()
```

```python
# Train, Test1, Test2 Data 저장
df_concat_1117.to_csv('./datasets/KOSDAQ+KOSPI_TRAIN_DATA.csv')
df_concat_1819.to_csv('./datasets/KOSDAQ+KOSPI_TEST1_DATA.csv')
df_concat_2021.to_csv('./datasets/KOSDAQ+KOSPI_TEST2_DATA.csv')
```

# Kosdaq Data Spilt

```python
# 회계년도 기준 데이터 스플릿 진행
kosdaq_1117 = kosdaq[kosdaq['회계년도'] <= '2017/12']
kosdaq_1819 = kosdaq[(kosdaq['회계년도'] <= '2019/12') & (kosdaq['회계년도'] >= '2018/12')]
kosdaq_2021 = kosdaq[(kosdaq['회계년도'] <= '2021/12') & (kosdaq['회계년도'] >= '2020/12')]
```
